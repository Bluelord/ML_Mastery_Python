## Univairal Plots

To understand your data we review our fetures individualy as well as combine plots.
Some of the techniques are discribed bellow
*   **Histogram** grups the data into bars provide uss with the counts of a values for perticular fetuere, it gives observation how much skew and Gaussian or its distribution.
*   **Density** also gives an idea about the distribution, this is smooth for mof histogram.
*   **Box & Whiaker**  summarizes the distribution for our feature, box gives us 25%, 50%, 75% and whisker gives the spread of the data outside the box. 


```python
# Univeriate Histogram
import numpy as np
from matplotlib import pyplot 
from pandas import read_csv
filename = "/content/drive/MyDrive/Colab Notebooks/ML Mastery python/Dataset/pima-indians-diabetes.csv"
names = ['preg', 'plas','pres','skin','test','mass','pedi','age','class']
data = read_csv(filename, names=names)
data.hist()
pyplot.show()
```

```python
# Univarient Density Plot
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False) # plotting discribtion
pyplot.show()
```

```python
# Box & Whisker Plot
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False) # plotting discribtion
pyplot.show()
```

## **Multivariate Plots**

**Correlation** gives the relation between any two variables(Features), correlation may be positive(Both are in same direction) or nagative(Both are in opposite direction). 
Correlation between each and every features gives us a correlation matrix, this matrix is symmetrical.


```python
# Correlation Matrix

data = read_csv(filename, names=names)
correlations = data.corr()
# Plot Correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
# For Generic correlation matrix the code between ################## is not needed
##############################
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
###############################
pyplot.show()
```

```python
data = read_csv(filename, names=names)
correlations = data.corr()
# plot correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
pyplot.show()
```
